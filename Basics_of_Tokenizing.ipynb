```
with open('/tmp/bbc-text.csv','r') as f:
  reader = list(csv.DictReader(f)) # We are converting to a list straightaway
labels=[]
sentences=[]
for item in reader:
  labels.append(item['category'])
  sentences.append(item['text'])
tokenizer=Tokenizer()
tokenizer.fit_on_texts(sentences)
wi = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(sentences)
maxlen = max([len(item) for item in sequences])
padded=pad_sequences(sequences,maxlen=maxlen,padding='post')
print(padded.shape)
```

```
t=Tokenizer()
t.fit_on_texts(labels)
label_indices = t.word_index
label_sequences = t.texts_to_sequences(labels)
print(label_indices)
print(label_sequences)
```
